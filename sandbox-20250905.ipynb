```python
import numpy as np

# Define the data type (structure) for the array
dtype = np.dtype([('name', 'U10'), ('age', 'i4'), ('height', 'f4')])

# Create a structured array with 3 elements
temperature = np.array([('Alice', 25, 5.5), ('Bob', 30, 5.9), ('Charlie', 35, 6.1)], dtype=dtype)

# Access elements
print("Full structured array:")
print(temperature)

# Access specific field (e.g., names)
print("\nNames:")
print(temperature['name'])

# Access specific element (e.g., first person's age)
print("\nAge of first person:", temperature[0]['age'])
```

```python
import xarray as xr
import numpy as np

# Define the dimensions
layers = ['A', 'B', 'C']
rows = range(3)
columns = ['Col1', 'Col2', 'Col3', 'Col4']

# Create data for each column with different dtypes
data_col1 = np.random.rand(3, 3).astype(np.float32)  # float32
data_col2 = np.random.randint(0, 65535, (3, 3), dtype=np.uint16)  # uint16
data_col3 = np.random.randint(0, 255, (3, 3), dtype=np.uint8)  # uint8
data_col4 = np.random.randint(0, 255, (3, 3), dtype=np.uint8)  # uint8

# Combine into an xarray.Dataset
dataset = xr.Dataset(
    {
        'Col1': (['Layer', 'Row'], data_col1),
        'Col2': (['Layer', 'Row'], data_col2),
        'Col3': (['Layer', 'Row'], data_col3),
        'Col4': (['Layer', 'Row'], data_col4),
    },
    coords={
        'Layer': layers,
        'Row': rows,
    }
)

dataset
```

```python
import xarray as xr
import numpy as np

# Create a simple 3D array of data with labeled coordinates
temperature = np.random.rand(4, 3, 2)  # A 3D array with shape (4, 3, 2)
precipitation = np.random.rand(4, 3, 2)  # Another 3D array with the same shape
time = pd.date_range('2023-01-01', periods=4)
latitude = np.linspace(-90, 90, 3)
longitude = np.linspace(-180, 180, 2)

# Create an xarray Dataset
ds = xr.Dataset(
    {
        'temperature': (['time', 'latitude', 'longitude'], temperature),
        'precipitation': (['time', 'latitude', 'longitude'], precipitation)
    },
    coords={
        'time': time,
        'latitude': latitude,
        'longitude': longitude

    }
)

print(ds)
```

```python
import xarray as xr
import numpy as np

# Create a 4D array with different data types
temperature = np.random.rand(3, 4, 2, 5)  # 4D array of shape (3, 4, 2, 5)

# Create an xarray Dataset with 4D data
ds = xr.Dataset(
    {
        'var1': (['dim1', 'dim2', 'dim3', 'dim4'], temperature),
        'var2': (['dim1', 'dim2', 'dim3', 'dim4'], np.random.rand(3, 4, 2, 5))
    },
    coords={
        'dim1': ['A', 'B', 'C'],
        'dim2': ['X', 'Y', 'Z', 'W'],
        'dim3': ['one', 'two'],
        'dim4': [1, 2, 3, 4, 5]
    }
)

print(ds)
```
